```python
# Homework PyBank Budget Analysis

# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path('Resources/budget_data.csv')

# Initialize line_num variable
line_num = 0

# Initialize metric variables

max_profit = 0
max_loss = 0
avg_profit = 0
total_profit = 0
count_month = 0
max_profit_date = None
max_loss_date = None

# Open the input path as a file object
with open(csvpath, 'r') as csvfile:

    # Print the datatype of the file object
    #print(type(csvfile))
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Print the datatype of the csvreader
    #print(type(csvreader))
    
    # Go to the next row from the start of the file
    # (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    line_num += 1
    # Print the header
    #print(f"{header} <---- HEADER")

    # Read each row of data after the header
    for row in csvreader:
        # Print the row
        #print(row)
        
        #The total number of months included in the dataset
        count_month += 1
        
        #The net total amount of Profit/Losses over the entire period. 
        total_profit += int(row[1])
        
        #The greatest increase/decrease in profits over the entire period.
        if max_loss == 0:
            max_loss = int(row[1])
        elif int(row[1]) < max_loss:
            max_loss = int(row[1])
            max_loss_date = row[0]
        elif int(row[1]) > max_profit:
            max_profit = int(row[1])
            max_profit_date = row[0]
            
#The average of the changes in Profit/Losses over the entire period.

avg_profit = round(total_profit/count_month,2) 

# set output path

output_path = Path('output.txt')

# Open the output path as a file object and write the output in txt file

with open(output_path, 'w') as txtfile:
    print("      Financial Analysis     ",file=txtfile)
    print("-----------------------------",file=txtfile)
    print(f"Total Months: {count_month}",file=txtfile)
    print(f"Total: ${total_profit}",file=txtfile)
    print(f"Average  Change: ${avg_profit}",file=txtfile)
    print(f"Greatest Increase in Profits:{max_profit_date} (${max_profit})",file=txtfile)
    print(f"Greatest Decrease in Profits:{max_loss_date} (${max_loss})",file=txtfile)

#printing output in terminal

with open(output_path, 'r') as txtfile:
    print(txtfile.read())
```
